## Create & Access SQLite database using Python

### Task 1: Create database using SQLite


```python
#Install & load sqlite3

#!pip install sqlite3  ##Uncomment this code only if you are working in a local environment to install sqlite3

import sqlite3
```

```python
# Connecting to sqlite
# connection object

conn = sqlite3.connect('INSTRUCTOR.db')
```

```python
## cursor object

cursor_obj =conn.cursor()
```

### Task 2: Create a table in the database

```python
# Drop the table if already exists.

cursor_obj.execute("DROP TABLE IF EXISTS INSTRUCTOR")
```

```python
# Create table

table ="""create table IF NOT EXISTS INSTRUCTOR(ID INTEGER PRIMARY KEY NOT NULL, FNAME VARCHAR(20), LNAME VARCHAR(20), CITY VARCHAR(20), CODE VARCHAR(20));"""

cursor_obj.execute(table)

print("Table is ready")
```

## Task 3: Insert data into the table

```python
cursor_obj.execute('''insert into INSTRUCTOR values (1, 'Rav', 'Ahuja', 'TORONTO', 'CA')''')
```

```python
cursor_obj.execute('''insert into INSTRUCTOR values (2, 'Raul', 'Chong', 'Markham', 'CA'), (3, 'Hima', 'Vasudevan', 'Chicago', 'US')''')
```

### Task 4: Query data in the table

```python
statment = '''SELECT *  FROM INSTRUCTOR'''

cursor_obj.execute(statment)

print("All the data")

output_all = cursor_obj.fetchall()

for row_all in output_all:
    print(row_all)
```

```python
## fetch few rows from data

statment = '''SELECT * FROM INSTRUCTOR'''
cursor_obj.execute(statment)

print('All the data')
# If you want to fetch few rows from the table we use fetchmany(numberofrows) and mention the number how many rows you want to fetch
output_all = cursor_obj.fetchmany(2)

for row_many in output_all:
    print(row_many)
```

```python
## Fetch FNAME from data

statment = '''SELECT FNAME FROM INSTRUCTOR'''
cursor_obj.execute(statment)

print("All the data")
output_column = cursor_obj.fetchall()
for fetch in output_column:
    print(fetch)
```

### update statement

```python
# execute an update statement that changes the Rav's CITY to MOOSETOWN

query_update = '''update INSTRUCTOR set CITY= 'MOOSETOWN' WHERE FNAME= 'Rav' '''
cursor_obj.execute(query_update)

statement = '''SELECT * FROM INSTRUCTOR'''
cursor_obj.execute(statement)

print("All the data")

output1 = cursor_obj.fetchmany(2)
for row in output1:
    print(row)
```
